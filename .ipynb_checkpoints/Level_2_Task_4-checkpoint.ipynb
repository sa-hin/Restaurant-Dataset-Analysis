## Restaurant Chains

```python
df = pd.read_csv('Dataset .csv')
df
```

## Identify if there are any restaurant chains present in the dataset

```python
from IPython.display import display
# Grouped by restaurant names and count occurrences
restaurant_counts = df['Restaurant Name'].value_counts()

# Identified restaurant chains (restaurants appearing more than once)
restaurant_chains = restaurant_counts[restaurant_counts > 1]

print("Restaurant chains identified:")
print(restaurant_chains)  
```

```python
# Filtered the data to include only these chains
chains_df = df[df['Restaurant Name'].isin(restaurant_chains.index)]


# Visualized the results
import folium

# Created a base map centered around the mean coordinates
mean_lat = chains_df['Latitude'].mean()
mean_lng = chains_df['Longitude'].mean()
base_map = folium.Map(location=[mean_lat, mean_lng], zoom_start=12)

# Add points to the map, coloring by restaurant chain
for idx, row in chains_df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(base_map)

# Saved the map to an HTML file (optional)
base_map.save('restaurant_chains_map.html')

# Displayed the map in a Jupyter Notebook (if using one)
display(base_map) 
```

## Analyze the ratings and popularity of different restaurant chains

```python
restaurant_counts = df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1].index

# Filtered the data to include only these chains
chains_df = df[df['Restaurant Name'].isin(restaurant_chains)]

# Calculated average rating and total votes for each chain
chain_stats = chains_df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).reset_index()

# Renamed columns for clarity
chain_stats.columns = ['Restaurant Name', 'Average Rating', 'Total Votes']

# Sorted by average rating and select top 10
top_10_chains = chain_stats.sort_values(by='Average Rating', ascending=False).head(10)

# Printed the top 5 chain statistics
print("Top 5 Restaurant Chains by Average Rating:")
print(top_10_chains)
```

```python
# Visualized the results
# Ploted average rating of top 10 restaurant chains
plt.figure(figsize=(12, 6))
plt.bar(top_10_chains['Restaurant Name'], top_10_chains['Average Rating'], color='skyblue')
plt.xlabel('Restaurant Chain')
plt.ylabel('Average Rating')
plt.title('Top 5 Restaurant Chains by Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ploted the popularity (total votes) for these top 10 chains

plt.figure(figsize=(12, 6))
plt.bar(top_10_chains['Restaurant Name'], top_10_chains['Total Votes'], color='salmon')
plt.xlabel('Restaurant Chain')
plt.ylabel('Total Votes')
plt.title('Popularity of Top 10 Restaurant Chains (Total Votes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show() 
```
