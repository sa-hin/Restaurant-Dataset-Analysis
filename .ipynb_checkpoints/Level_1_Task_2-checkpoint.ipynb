### Identify the city with the highest number of restaurants in the dataset

```python
resturant_count = df['City'].value_counts()
resturant_count 
```

```python
max_resturant_count = resturant_count.max()
max_resturant_count
```

```python
max_resturant_city = resturant_count.idxmax()
max_resturant_city 
```

```python
print(f"The city with the highest number of resturants is '{max_resturant_city}' with {max_resturant_count} resturants.") 
```

## Calculate the average rating for restaurants in each city

```python
average_rating_per_city = df.groupby('City')['Aggregate rating'].mean().reset_index() 
average_rating_per_city
```

```python
average_rating_per_city_sorted = average_rating_per_city.sort_values(by = 'Aggregate rating', ascending = False).head(5)
average_rating_per_city_sorted
```

```python
colors = plt.cm.tab20b(range(len(average_rating_per_city_sorted)))

plt.figure(figsize=(12,6))

bars = plt.bar(average_rating_per_city_sorted['City'], average_rating_per_city_sorted['Aggregate rating'], color= colors, width=0.6)
plt.title('Average Rating for Restaurants in Each City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()

# plt.legend(bars, average_rating_per_city_sorted['City'], loc='upper right') 
plt.show()
```

## Determine the city with the highest average rating

```python
top_rating_city = average_rating_per_city_sorted.iloc[0]
print(f"The highest average rating city is {top_rating_city['City']} with an average rating of {top_rating_city['Aggregate rating']}")
```
