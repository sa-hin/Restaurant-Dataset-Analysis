## Determine the percentage of restaurants that offer online delivery

```python
delivery_df= pd.Series({'No': 7100, "Yes": 2451})
online_delivery_percentage = (delivery_df['Yes']/ delivery_df.sum()) * 100
online_delivery_percentage.round(2)
```

## Compare the average ratings of restaurants with and without online delivery

```python
average_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()
average_ratings
```

```python
# Visualized the comparison with a bar chart
labels = average_ratings.index
ratings = average_ratings.values
colors = ['skyblue', 'salmon']

plt.figure(figsize=(6, 4))
plt.bar(labels, ratings, color=colors, width=0.6)
plt.title('Average Ratings of Restaurants With and Without Online Delivery')
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.ylim(0, 5)  # Assuming the rating scale is from 0 to 5
plt.tight_layout()
plt.show()
```
