## Identify the most common combinations of cuisines in the dataset.

```python
df['Cuisines'].isnull().sum()
```

```python
cuisines = df['Cuisines'].dropna()
cuisines
```

```python
cuisine_combinations = cuisines.apply(lambda x: frozenset(x.split(',')))

# Counted the occurrences of each unique combination
combination_counts = cuisine_combinations.value_counts()

# Printed the most common combinations
print("The most common combinations of cuisines are:")
print(combination_counts.head())
```

## Determine if certain cuisine combinations tend to have higher ratings

```python
import warnings 
warnings.filterwarnings("ignore")
df = df.dropna(subset=['Cuisines', 'Aggregate rating'])

df['Cuisine Combination'] = df['Cuisines'].apply(lambda x: frozenset(x.split(',')))
average_ratings_by_combination = df.groupby('Cuisine Combination')['Aggregate rating'].mean()
sorted_average_ratings = average_ratings_by_combination.sort_values(ascending=False)

print("Cuisine combinations with the highest average ratings:")
print(sorted_average_ratings.head(10))

```

```python
# Visualized the top 10 combinations with a bar chart
import matplotlib.pyplot as plt

top_combinations = sorted_average_ratings.head(10)
top_combinations.plot(kind='bar', figsize=(12, 6), color='skyblue', edgecolor='black')
plt.title('Top 10 Cuisine Combinations by Average Rating')
plt.xlabel('Cuisine Combination')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
